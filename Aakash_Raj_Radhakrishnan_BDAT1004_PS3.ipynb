````python
# Step 1: Import the necessary libraries
import pandas as pd
````

````python
# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
````

````python
# Step 3: Data is now assigned to the 'users' variable
users = pd.read_csv(url, sep='|')
users.info()
````

````python
# Step 4: Mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
mean_age_per_occupation
````

````python
# Step 5: Male ratio per occupation
def male_ratio(series):
    return (series == 'M').sum() / series.count()
male_ratio_per_occupation = users.groupby('occupation')['gender'].apply(male_ratio)
male_ratio_per_occupation = male_ratio_per_occupation.sort_values(ascending=False)
male_ratio_per_occupation
````

````python
# Step 6: Minimum and maximum ages per occupation
min_age_per_occupation = users.groupby('occupation')['age'].min()
min_age_per_occupation
````

````python
max_age_per_occupation = users.groupby('occupation')['age'].max()
max_age_per_occupation
````

````python
# Step 7: Mean age per occupation and sex
mean_age_per_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean()
mean_age_per_occupation_sex
````

````python
# Step 8: Percentage of women and men per occupation
def gender_percentage(series):
    total = series.count()
    return (series == 'F').sum() / total * 100, (series == 'M').sum() / total * 100
gender_percentage_per_occupation = users.groupby('occupation')['gender'].apply(gender_percentage)
female_percentage_per_occupation = gender_percentage_per_occupation.apply(lambda x: x[0])
male_percentage_per_occupation = gender_percentage_per_occupation.apply(lambda x: x[1])
````

Step 1. Import the necessary libraries

Step 2. Import the dataset from the attached file wind.txt

Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

````python
# Step 1: Import the necessary libraries
import pandas as pd
import numpy as np
import datetime
````

````python
# Step 2: Import the dataset
data = pd.read_csv("wind.txt", delim_whitespace=True, parse_dates=[[0, 1, 2]])
````

````python
# Step 3: Set datetime
data
````

````python
# Step 4: Function to fix the year 2061
def century(x):
  year = x.year - 100 if x.year > 1989 else x.year
  return datetime.date(year, x.month, x.day)
````

````python
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(century)
data
````

````python
# Step 5: Convert the index to datetime64
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])
data = data.set_index('Yr_Mo_Dy')
data
````

````python
# Step 6: Count missing values for each location
missing_values_per_location = data.isnull().sum()
missing_values_per_location
````

````python
# Step 7: Count non-missing values in total
non_missing_values = data.notnull().sum()
non_missing_values
````

````python
total_non_missing_values = data.count().sum()
total_non_missing_values
````

````python
# Step 8: Calculate the mean windspeed over all locations and times
mean_windspeed = data.stack().mean()
mean_windspeed
````

````python
# Step 9: Calculate statistics for each location
loc_stats = data.describe()
loc_stats
````

````python
# Step 10: Calculate statistics for each day
day_stats = data.resample('D').agg(['min', 'max', 'mean', 'std'])
day_stats
````

````python
day_stats = pd.DataFrame()
day_stats['min'] = data.min(axis = 1)
day_stats['max'] = data.max(axis = 1)
day_stats['mean'] = data.mean(axis = 1)
day_stats['std'] = data.std(axis = 1)

day_stats
````

````python
# Step 11: Average windspeed in January for each location
average_january_windspeed = data.loc[data.index.month == 1].mean()
average_january_windspeed
````

````python
# Step 12: Downsample to yearly frequency
yearly_resampled = data.groupby(data.index.to_period('Y')).mean()
yearly_resampled
````

````python
# Step 13: Downsample to monthly frequency
monthly_resampled = data.groupby(data.index.to_period('M')).mean()
monthly_resampled
````

````python
# Step 14: Downsample to weekly frequency
weekly_resampled = data.groupby(data.index.to_period('W')).mean()
weekly_resampled
````

````python
# Step 15: Calculate statistics for the first 52 weeks
weekly_stats = data.resample('W').agg(['min','max','mean','std'])
weekly_stats.loc[weekly_stats.index[1:53], "RPT":"MAL"]
````

# Question 6

Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.

Don't forget to label your axes!


````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Load the dataset
data = pd.read_csv("us-marriages-divorces-1867-2014.csv")
````

````python
# Calculate marriages and divorces per capita
data['Marriages_per_capita'] = data['Marriages'] / data['Population'] * 1000
data['Divorces_per_capita'] = data['Divorces'] / data['Population'] * 1000
````

````python
# Set ggplot style
plt.style.use('ggplot')
````

````python
# Line plot
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Marriages_per_capita'], label='Marriages per capita')
plt.plot(data['Year'], data['Divorces_per_capita'], label='Divorces per capita')
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.legend()
plt.grid()
plt.show()
````

# Question 8

Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name.

Don't forget to label your axes!


````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Load the dataset
data = pd.read_csv("actor_kill_counts.csv")
````

````python
# Sort the data by kill count in descending order
data = data.sort_values(by='Count', ascending=True)
````

````python
# Set ggplot style
plt.style.use('ggplot')
````

````python
# Horizontal bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(data['Actor'], data['Count'])

# Add the kill count at the end of each bar
for bar in bars:
    plt.text(bar.get_width() + 10, bar.get_y() + bar.get_height() / 2, str(int(bar.get_width())), va='center')

plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')
plt.grid(axis='x')
plt.show()
````

# Question 9

Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.


Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Load the dataset
data = pd.read_csv("roman-emperor-reigns.csv")
````

````python
# Count the number of emperors that were assassinated
assassinated_count = data[data['Cause_of_Death'] == 'Assassinated'].shape[0]
````

````python
# Count the number of emperors that were not assassinated
not_assassinated_count = data.shape[0] - assassinated_count
````

````python
# Create a pie chart
labels = ['Assassinated', 'Not Assassinated']
sizes = [assassinated_count, not_assassinated_count]
colors = ['#FF6666', '#99FF99']
explode = (0.1, 0)

plt.figure(figsize=(6, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')

plt.title('Fraction of Roman Emperors that were Assassinated')
plt.show()
````
